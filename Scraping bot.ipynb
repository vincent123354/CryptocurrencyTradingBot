```python
import numpy as np
```

```python
total_quantities = np.arange(0.001, 100, step=0.001)
history = []

for total_quantity in total_quantities:
    total_order = 0
    minimum_entry = 0.001
    tp_quantity = [0.2, 0.7, 0.1]
    tmp_total_quantity = 0
    for i in range(len(tp_quantity)):
        if i == len(tp_quantity) - 1:
            if total_quantity - tmp_total_quantity == 0:
                print(total_quantity, tmp_total_quantity)
            total_order += total_quantity - tmp_total_quantity
            break
            
        precision = 3
        quantity = round(max(minimum_entry, total_quantity * tp_quantity[i]), precision)

        total_order += quantity
        tmp_total_quantity += quantity
        if tmp_total_quantity >= total_quantity:
            break

#         print('Open order', quantity)
    history.append(total_order == total_quantity)
```
